```python
import pandas as pd
districts=pd.read_csv('/content/district_wise.csv')
districts=districts[districts['State']=='Uttar Pradesh']
districts['District']=districts['District'].str.lower()
```

```python
districts=districts.replace('bhadohi','sant ravidas nagar (bhadohi)')
districts=districts.replace('maharajganj','mahrajganj')
districts=districts.replace('barabanki','bara banki')
districts=districts.replace('shamli','samli')
districts=districts.replace('hathras','mahamaya nagar')
districts=districts.replace('ayodhya','faizabad')
districts=districts.replace('lakhimpur kheri','kheri')
districts=districts.replace('prayagraj','allahabad')
districts=districts.replace('kasganj','kanshiram nagar')
districts=districts.replace('amroha','jyotiba phule nagar')
```

```python
districts=districts.drop(['State','SlNo','State_Code','District_Key','Migrated_Other','Delta_Confirmed','Delta_Active','Delta_Deceased','Delta_Recovered','District_Notes','Last_Updated'],axis=1)
```

```python
districts['District'].unique()
```

```python
districts=districts.fillna(0)
```
