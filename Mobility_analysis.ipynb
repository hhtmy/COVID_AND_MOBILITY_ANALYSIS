```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/lucknow population growth rate data - Sheet1.csv')
```

```python
import matplotlib.cm as cm
import numpy as np
plt.figure(figsize=(15, 10))
x=df['Year']
y=df['Growth Rate(%)']
plt.xticks(rotation = '45')
plt.title('Growth in Population density of Lucknow')
plt.xlabel('YEAR')
plt.ylabel('Population Density')
plt.bar(x,y,
        color = cm.rainbow(np.linspace(0, 1, len(y))))
```

```python

x=df['Year']
y=df['Growth Rate(%)']
plt.plot(x,  y, 
             linewidth = 4)
plt.title('Variation of Population density in Lucknow')
plt.xlabel('Years')
plt.xticks(rotation = '90')
plt.ylabel('Population Density')
plt.legend(prop = {'size': 24})
```
